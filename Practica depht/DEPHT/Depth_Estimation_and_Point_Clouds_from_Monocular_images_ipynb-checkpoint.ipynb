```python
import cv2
import os
import torch
import urllib.request
import numpy as np
import matplotlib.pyplot as plt
plasma = plt.get_cmap('plasma')
url, filename = ("https://github.com/pytorch/hub/raw/master/images/dog.jpg", "dog.jpg")
urllib.request.urlretrieve(url, filename)
```

```python
model_name = 'DPT_Hybrid'
model = torch.hub.load("intel-isl/MiDaS", model_name)
```

```python
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
model.to(device)
model.eval()
```

```python
midas_transforms = torch.hub.load("intel-isl/MiDaS", "transforms")
transform = midas_transforms.dpt_transform
```

```python
def pipeline(img_path):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    input_batch = transform(img).to(device)
    with torch.no_grad():
        prediction = model(input_batch)

        prediction = torch.nn.functional.interpolate(
            prediction.unsqueeze(1),
            size=img.shape[:2],
            mode="bicubic",
            align_corners=False,
        ).squeeze()

    output = prediction.cpu().numpy()
    return output
```

```python
def rescale_output(pred):
    min = np.min(pred)
    max = np.max(pred)
    return (pred-min)/max
```

```python
def get_img_array(filename):
    im = cv2.imread(filename)
    im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    x = np.clip(np.asarray(im, dtype=float) / 255, 0, 1)
    return x
```

```python
folder_path = '/home/kevin/Documentos/SEPTIMO_SEMESTRE/TESIS/Imagenes'
for filename in os.listdir(folder_path):
    filename = os.path.join(folder_path, filename)
    ## Run Depth Estimation on input filename
    output = pipeline(filename)
    ## Rescale depth output to relative depth
    output = rescale_output(output)
    depth_orig = output.copy()
    ## Color Depth output on plasma colormap
    colored_depth = plasma(output)[:, :, :3]
    ## Stack input image and depth prediction
    imgs = []
    input_img_array =get_img_array(filename)
    print(input_img_array.shape)
    imgs.append(input_img_array)
    imgs.append(colored_depth)
    img_set = np.hstack(imgs)
    plt.figure(figsize = (20,10))
    plt.imshow(img_set)
    plt.show()
    break
```

```python
from kornia.geometry.depth import depth_to_3d
import open3d as o3d
```

```python
def get_3d_point_cloud(depth_map):
    height, width = depth_map.shape[:2]
    camera_matrix = torch.FloatTensor([[587, 0, width/2],
                                       [0, 587, height/2],
                                        [0, 0, -1]]).unsqueeze(0)
    depth_map_min = np.min(depth_map)
    depth_map_max = np.max(depth_map)
    scaled_depth = (depth_map - depth_map_min)/depth_map_max
    scaled_depth = torch.from_numpy(scaled_depth).view(1, 1, height, width)
    point_clouds = depth_to_3d(
        scaled_depth, camera_matrix, normalize_points=False)
    cloud_room = point_clouds.permute(0, 2, 3, 1)[0].numpy()
    return cloud_room
```

```python
point_clouds = get_3d_point_cloud(depth_orig)
print(point_clouds.shape)
```

```python
pcd_room = point_clouds.reshape(-1,3)
print(pcd_room.shape)
```

```python
pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(pcd_room)
o3d.io.write_point_cloud("chair.ply", pcd)
```
